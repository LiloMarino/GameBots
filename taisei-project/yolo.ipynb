```python
from ultralytics import YOLO

# Carregar modelo pré-treinado (YOLOv5 small)
model = YOLO("yolov5s.pt")

# Treinar
model.train(
    data="../../LabelStudio/dataset/data.yaml",  # dataset
    epochs=50,        # menos épocas p/ não pesar tanto (pode subir depois)
    imgsz=640,        # 640 é padrão, mas se travar pode usar 416
    batch=8,          # controla consumo de memória. Se travar, tente 4 ou até 2
    workers=0,        # evita sobrecarregar CPU em máquinas pessoais
    device=0,         # usa GPU se tiver (0 = CUDA:0, "cpu" = força CPU)
    patience=10,      # early stopping automático se não melhorar
    save=True,        # salva melhor modelo
    pretrained=True   # aproveita pesos pré-treinados
)
```
